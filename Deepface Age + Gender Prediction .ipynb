```python
# import libraries
import pandas as pd
import tweepy
from deepface import DeepFace
import time
```

```python
# the user_df will contain all of the author user_ids 
user_df = pd.read_csv("authors.csv", header = None, names = ['user_id'])
```

```python
user_df.head()
```

```python
# creating and exporting a dataframe that contains every user id and their corresponding profile picture link

# initialize api
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""

#Pass in our twitter API authentication key
auth = tweepy.OAuth1UserHandler(
    consumer_key, consumer_secret,
    access_token, access_token_secret
)

#Instantiate the tweepy API
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
len(user_df)
```

```python
user_df.drop_duplicates(inplace = True)
```

```python
len(user_df)
```

```python
cols = ["id", "Profile Picture", "bio"]
df = pd.DataFrame(columns = cols)
```

```python
df
```

```python
len(user_df)
```

```python
for count, user_id in enumerate(user_df['user_id']):
    user_id = user_id
    try:
        #The number of tweets we want to retrieved from the user
        user = api.get_user(user_id=user_id)

        #Pulling Some attributes from the tweet
        attributes_container = [user.id, user.profile_image_url, user.description]

        #Creation of Dataframe
        df.loc[count] = attributes_container
        
    except BaseException as e:
        #print('Status Failed On,',str(e))
        df.loc[count] = [user_id, str(e), str(e)]
        time.sleep(3)
        continue
```

```python
final_df = pd.DataFrame()
final_df['user_id']=[]
final_df['gender']=[]
final_df['age'] = []
```

```python
####

#final_df.index = df['id']

for user_id, picture_link in zip(df['id'], df['Profile Pictue']):
    
    # need to remove '_normal' in order to enhance jpeg
    img = picture_link.replace("_normal", "")
    
    try:
        results = DeepFace.analyze(img, actions= ("gender", "age"), enforce_detection = True)
        final_df.loc[user_id] = [user_id, results[0]['dominant_gender'], results[0]['age']]
    except:
        final_df.loc[user_id] = [user_id, None, None]
        continue
        
final_df.head(10)
```
