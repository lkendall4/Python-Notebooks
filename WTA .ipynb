```python
def namesMatch(x,y):
    if x == y:
        return True
    else:
        return False
```

```python
final_df=pd.DataFrame()
for i, x in zip(df_list,df_list_names):
    for y, z in zip(df_list, df_list_names):
        if namesMatch(x,z):
            continue
        temp = pd.merge(i,y,how='inner', on=['Tournament', 'Surface','Time','Score', 'Rd', 'Date'])
        final_df= pd.concat([final_df,temp])
```

```python
def remove_rk (df, column_name):
    column=[]
    for i in df[column_name]:
        if '10' not in i:
            i = i[3:]
            column.append(i)
        else:
            i = i[4:]
            column.append(i)
    df[column_name] = column
```

```python
# I want it to be w_1st% instead of player y and x. Need to see who won, move their values to new columns with new names

final_df['w_df%'] = final_df['DF%_x']
final_df['l_df%'] = final_df['DF%_y']
final_df['w_rk%'] = final_df['Rk_x']
final_df['l_rk%'] = final_df['Rk_y']
final_df['w_1stIn'] = final_df['1stIn_x']
final_df['l_1stIn'] = final_df['1stIn_y']
final_df['w_1st%'] = final_df['1st%_x']
final_df['l_1st%'] = final_df['1st%_y']
final_df['w_A%'] = final_df['A%_x']
final_df['l_A%'] = final_df['A%_y']
final_df['w_2nd%'] = final_df['2nd%_x']
final_df['l_2nd%'] = final_df['2nd%_y']

for i,j in zip(final_df['winner'], final_df.index):
    if final_df.iloc[j, 31] == i: # then player y won, data needs to be switched around
        
        y_rank = final_df.iloc[j,20] #yrank
        final_df.at[j, 'w_rk'] = y_rank
        x_rank = final_df.iloc[j,5] #xrank
        final_df.at[j,'l_rk'] = x_rank
        
        y_df_percent = final_df.iloc[j,26] #y_df%
        final_df.at[j, 'w_df%'] = y_df_percent
        x_df_percent = final_df.iloc[j,12] #x
        final_df.at[j,'l_df%'] = x_df_percent
        
        y_first_in = final_df.iloc[j,27] 
        final_df.at[j, 'w_1stIn'] = y_first_in
        x_first_in = final_df.iloc[j,13] 
        final_df.at[j,'l_1stIn'] = x_first_in
        
        y_first_percent = final_df.iloc[j,28] 
        final_df.at[j, 'w_1st%'] = y_first_percent
        x_first_percent = final_df.iloc[j,14] 
        final_df.at[j,'l_1st%'] = x_first_percent
        
        y_ace = final_df.iloc[j,25] 
        final_df.at[j, 'w_A%'] = y_ace
        x_ace = final_df.iloc[j,11] 
        final_df.at[j,'l_A%'] = x_ace
        
        y_second_percent = final_df.iloc[j,29] 
        final_df.at[j, 'w_2nd%'] = y_second_percent
        x_second_percent = final_df.iloc[j,15] 
        final_df.at[j,'l_2nd%'] = x_second_percent
```
